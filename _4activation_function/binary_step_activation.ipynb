```python
# binary step activation function
def binary_step(x):
    if x < 0:
        return 0
    else:
        return 1
```

```python
binary_step(-0.7)
```

```python
import numpy as np

input = np.arange(-7,7,0.15)
print(input.shape)
print(input[:10])
print(input[-11:-1])

output = np.array([binary_step(x) for x in input])
print(output[40:55])
```

```python
import matplotlib.pyplot as plt

plt.plot(input, output)
plt.xlabel('x (Input)')
plt.ylabel('y (Output)')
```
