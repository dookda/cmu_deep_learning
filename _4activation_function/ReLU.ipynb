```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ReLU function
def ReLU(x):
    return np.maximum(0,x)
```

```python
ReLU(-1), ReLU(0), ReLU(0.5), ReLU(1), ReLU(2)
```

```python
input = np.arange(-5.0, 5.0, 0.1)
output = np.array([ReLU(x) for x in input])

plt.ylabel("output(y)")
plt.xlabel("input(x)")
plt.grid(True, linestyle = "--", color = "gray", linewidth = "0.5", axis = "both", which = "both")
plt.plot(input, output, color = "blue", linewidth = "2")
```
