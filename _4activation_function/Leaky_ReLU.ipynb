```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leaky ReLU function
def leaky_relu(x):
    return np.maximum(0.1*x, x)
```

```python
input = np.arange(-5, 5, 0.01)
output = np.array([leaky_relu(x) for x in input])

plt.ylabel("output(y)")
plt.xlabel("input(x)")
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.plot(input, output)
```
