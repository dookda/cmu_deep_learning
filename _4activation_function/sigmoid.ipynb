```python
from math import exp

def sigmoid(x):
    return 1 / (1 + exp(-x))
```

```python
sigmoid(0),sigmoid(1),sigmoid(2),sigmoid(3),sigmoid(6),sigmoid(10),sigmoid(50),sigmoid(500),sigmoid(-8)
```

```python
import numpy as np
import matplotlib.pyplot as plt

input = np.arange(-8,8,0.1)
output = np.array([sigmoid(x) for x in input])
plt.axhline(y=0.5, color='r', linestyle='--', linewidth=0.8, label='y=0.5')
plt.ylabel('y (output)')
plt.xlabel('x (input)')
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, color='gray')
plt.plot(input,output, c='b')
```
