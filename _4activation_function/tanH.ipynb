```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp

# tanh function
def tanh(x):
    return (exp(x) - exp(-x)) / (exp(x) + exp(-x))

# tanh derivative
def tanh_derivative(x):
    return 1 - tanh(x)**2
```

```python
tanh(0),tanh(1),tanh(2),tanh(3),tanh(4),tanh(50),tanh(100),tanh(-1),tanh(-8),tanh(-50)
```

```python
input = np.arange(-10, 10, 0.01)
output = np.array([tanh(x) for x in input])

plt.ylabel('y (output)')
plt.xlabel('x (input)')
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, color='gray')
plt.plot(input, output, color='g',lw=2)
```
