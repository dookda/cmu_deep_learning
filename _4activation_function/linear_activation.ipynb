```python
# linear activation function
def linear_05(x):
    return 0.5 * x

def linear_1(x):
    return x

def linear_2(x):
    return 2 * x
```

```python
linear_05(2), linear_1(2), linear_2(2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

input = np.arange(-10, 10, 0.15)
output05 = np.array([linear_05(x) for x in input])
output1 = np.array([linear_1(x) for x in input])
output2 = np.array([linear_2(x) for x in input])

# print(output05)
plt.plot(input, output05, lw=2, c="g" ,label='linear_05')
plt.plot(input, output1, lw=2, c="b" ,label='linear_1')
plt.plot(input, output2, lw=2, c="r" ,label='linear_2')
plt.ylabel('Output')
plt.xlabel('Input')
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, color='gray')
plt.legend()
```
